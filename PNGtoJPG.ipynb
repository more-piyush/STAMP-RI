```python
import csv
import shutil
import os
import sys

import os
from PIL import Image

# Path where PNG images are located
input_folder = './positive_images'  # Change this to your folder path if needed
output_folder = './messidor'
os.makedirs(output_folder, exist_ok=True)
# Get list of PNG files
png_files = [file for file in os.listdir(input_folder) if file.lower().endswith('.png')]
png_files.sort()  # Optional: sort for consistent ordering

# Convert and save as JPG with new names
for idx, file_name in enumerate(png_files, start=1):
    png_path = os.path.join(input_folder, file_name)
    jpg_name = f'c{idx}.jpg'
    jpg_path = os.path.join(output_folder, jpg_name)

    with Image.open(png_path) as img:
        rgb_img = img.convert('RGB')  # PNG can have alpha, JPG can't
        rgb_img.save(jpg_path, 'JPEG')

    print(f"Converted: {file_name} -> {jpg_name}")
```
